```python
import timm
import torch.nn as nn

model = timm.create_model("hf_hub:cm93/resnet18-eurosat", pretrained=True)

print(model)
```

```python
class ResNet18(nn.Module):
    def __init__(self):
        super(ResNet18, self).__init__()
        self.model = timm.create_model("hf_hub:cm93/resnet18-eurosat", pretrained=True)
        self.model.fc = nn.Linear(self.model.fc.in_features, 11255)
        self.model.softmax = nn.Softmax(dim=1)
        
    def forward(self, x):
        output = self.model(x)
        output = self.model.softmax(output)
        return self.model(x)

model = ResNet18()
print(model)
```
